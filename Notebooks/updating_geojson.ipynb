```python
import pandas as pd
```

```python
# Read CSV files into pandas DataFrames
geojson_data = pd.read_csv("geojson_data.csv")
percent_data = pd.read_csv("Dans things/percent_data.csv")
```

```python
geojson_data
```

```python
percent_data
```

```python
# Merge DataFrames based on the "name" column
merged_data = pd.merge(geojson_data, percent_data[['name', 'HA Percent', 'HAProperties', 'OtherProperties']], on='name')
```

```python
# Keep only the necessary columns
merged_data = merged_data[['geometry_type', 'geometry_coordinates', 'name', 'description', 'HA Percent', 'HAProperties', 'OtherProperties']]
```

```python
merged_data
```

```python
merged_data = merged_data.dropna()
```

```python
merged_data.dtypes
```

```python
# Save the merged DataFrame to a new CSV file
merged_data.to_csv("merged_geojson_data.csv", index=False)
```

```python
import json
```

```python

# Convert DataFrame to GeoJSON feature collection
feature_collection = {
    "type": "FeatureCollection",
    "features": []
}

for index, row in merged_data.iterrows():
    feature = {
        "type": "Feature",
        "geometry": {
            "type": row['geometry_type'],
            "coordinates": json.loads(row['geometry_coordinates'])
        },
        "properties": {
            "name": row['name'],
            "description": row['description'],
            "HA_Percentage": row['HA Percent'],
            "HA_Properties": row['HAProperties'],
            "Other_Properties": row['OtherProperties']
        }
    }
    feature_collection['features'].append(feature)

```

```python
feature_collection
```

```python
# Save GeoJSON to file
with open("merged_data.geojson", "w") as geojson_file:
    json.dump(feature_collection, geojson_file)
```
